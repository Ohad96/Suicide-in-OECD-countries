### Performing import and reading of a database.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/idanyacobi-hub/Suicide-in-the-OECD-countries/main/Suicide%20in%20the%20OECD%20countries.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.rename(columns = {"gdp_for_year($)":'gdp_for_year', 'gdp_per_capita($)':'gdp_per_capita','suicides/100kpop':'suicides/100k_pop','HDIforyear':'HDI_for_year'}, inplace = True)
```

```python
df.info()
```

```python
df['suicides/100k_pop'] = df['suicides/100k_pop']/100000
```

```python
df
```

```python
OECD = ['Austria', 'Australia', 'Belgium', 'Canada', 'Chile', 'Colombia', 'Costa Rica', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Republic of Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Mexico', 'Netherlands', 'New Zealand', 'Norway', 'Poland', 'Portugal', 'Slovakia','Slovenia','Spain','Sweden','Switzerland','Turkey' ,'United Kingdom' ,'United States']
```

```python
df_OECD = df[df['country'].isin(OECD)]
```

 # $\color{navy}{\text{Suicide rates by OECD Countries}}$

## This graph shows the level of suicides in OECD countries in the years 1985-2016

```python
df_OECD_group = df_OECD.groupby('country')[['suicides/100k_pop']].mean().reset_index()
plt.figure(figsize=(8,32))
sns.barplot(data = df_OECD_group, y='country', x='suicides/100k_pop',order=df_OECD_group.sort_values('suicides/100k_pop',ascending = False).country)
plt.xticks(rotation=70)
plt.xlabel("Suicides percentage",fontsize=20)
plt.ylabel("Country",fontsize=20)
plt.title("Suicide percentage by country",fontsize=20)
plt.show()
```

 # $\color{navy}{\text{Suicides Trend by Gender}}$ 

## This graph shows the trend of suicides in OECD countries by women and mens

### It can be seen that the men's trend is in higher percentages, on the other hand, the women's trend is in low percentages and looks stable.

```python
fig, ax = plt.subplots(figsize=(10,7))
df_OECD.groupby(['year','sex'])[['suicides/100k_pop']].mean().unstack().plot(ax=ax)
plt.legend(['Female','Male'],fontsize=15)
plt.title('Suicides trend by gender',fontsize=15)
plt.show()
```

## We focused on the countries: Israel, Austria and Switzerland.

```python
df_israel = df[df['country']=='Israel']
df_austria= df[df['country']=='Austria']
df_switzerland = df[df['country']=='Switzerland']
```

```python
df2 = df[df['country'].isin(['Israel','Austria','Switzerland'])]
```

 # $\color{navy}{\text{Suicides presentage by country over years}}$


```python
df_Israel = df2[df2['country'] == "Israel"]
df_Austria = df2[df2['country'] == "Austria"]
df_Switzerland = df2[df2['country'] == "Switzerland"]
sns.lineplot(x=df_Israel['year'], y=df2['suicides/100k_pop'], marker="o", errorbar=None)
sns.lineplot(x=df_Austria['year'], y=df2['suicides/100k_pop'], marker="o", errorbar=None)
sns.lineplot(x=df_Switzerland['year'], y=df2['suicides/100k_pop'], marker="o", errorbar=None)
plt.legend(['Israel','Austria','Switzerland'])
plt.show()
```

## We found that there is no effect of conscription on the army and wars on suicide rates.
#### Israel has a mandatory conscription for the army and many wars but the percentage of suicides in it is the lowest
#### Compared to Austria, there is mandatory conscription for the army but there are not many wars
#### and in Switzerland there is no mandatory conscription for the army and there have never been wars, but still the percentage of those who suicides in them is higher

 # $\color{navy}{\text{Corelation of Suicides Rate with GDP per Capira}}$

```python



fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Corelation of Suicides rate with GDP per capita')

# df_switzerland
sns.regplot(ax=axes[0], x=df_switzerland['gdp_per_capita'], y=df_switzerland['suicides/100k_pop'],color='green',line_kws={"color": "black"})
axes[0].set_title('Switzerland')
axes[0].set(ylim=(0, 0.0005))

# austria
sns.regplot(ax=axes[1], x=df_austria['gdp_per_capita'], y=df_austria['suicides/100k_pop'],color='orange',line_kws={"color": "black"})
axes[1].set_title('Austria')
axes[1].set(ylim=(0, 0.0005))

# israel
sns.regplot(ax=axes[2], x=df_israel['gdp_per_capita'], y=df_israel['suicides/100k_pop'],color='blue',line_kws={"color": "black"})
axes[2].set_title('Israel')
axes[2].set(ylim=(0, 0.0005))
```

### We found that as the GDP increases, suicide rates decrease in the country.

 # $\color{navy}{\text{Corelation of Suicides Rate with HDI for year}}$

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Corelation of Suicides rate  with HDI rate')

# df_switzerland
sns.regplot(ax=axes[0], data=df_switzerland, x='HDI_for_year', y='suicides/100k_pop', color='green', line_kws={"color": "black"})

axes[0].set_title('Switzerland')
axes[0].set(ylim=(0, 0.0004))

# austria
sns.regplot(ax=axes[1], x=df_austria['HDI_for_year'], y=df_austria['suicides/100k_pop'],color='orange',line_kws={"color": "black"})
axes[1].set_title('Austria')
axes[0].set(ylim=(0, 0.0004))


# israel
sns.regplot(ax=axes[2], x=df_israel['HDI_for_year'], y=df_israel['suicides/100k_pop'],color='blue',line_kws={"color": "black"})
axes[2].set_title('Israel')
axes[0].set(ylim=(0, 0.0004))
```

### We found that as the HDI increases, suicide rates decrease in the country.
